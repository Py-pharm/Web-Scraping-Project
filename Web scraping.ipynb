```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
first_page_url = "https://hotels.ng/hotels-in-rivers/port-harcourt/1"
```

```python
pages = np.arange(1,42,1)  
```

```python
name = []
address = []
price_per_night = []
facilities = []
rating = []
no_of_reviews = []
likes = []
for page in pages:
    page = requests.get("https://hotels.ng/hotels-in-rivers/port-harcourt/"+str(page)+"")
    soup = BeautifulSoup(page.text, 'html.parser')
    hotel_info = soup.findAll('div', class_="listing-hotels")
    sleep(randint(2,8))
    for info in hotel_info:        
        hotel_name = info.find('h2', class_="listing-hotels-name").text        
        name.append(hotel_name)
        
        hotel_address = info.find('p', class_="listing-hotels-address color-dark hidden-xs hidden-sm").text        
        address.append(hotel_address)
        
        av_price_per_night = info.find('div', class_="col-xs-4 listing-hotels-prices" ).find('p', class_="listing-hotels-prices-discount").text.split("\n")[0].replace("₦","").replace(",","") if info.find('div', class_="col-xs-4 listing-hotels-prices" ).find('p', class_="listing-hotels-prices-discount") else "no price"      
        price_per_night.append(av_price_per_night)
        
        hotel_facilities = info.find('div', class_="listing-hotels-facilities hidden-xs").text.replace("\n\n\n", ",")        
        facilities.append(hotel_facilities)
        
        ratings = info.find('div', class_="listing-hotels-rating-box").find('span', class_="bkgrnd-v-good").text if info.find('div', class_="listing-hotels-rating-box").find('span', class_="bkgrnd-v-good") else "no rating"        
        rating.append(ratings)
        
        number_of_reviews = info.find('div', class_="listing-hotels-rating-box").find('p', class_="listing-hotels-rating-number").text.split(" ")[1] if info.find('div', class_="listing-hotels-rating-box").find('p', class_="listing-hotels-rating-number") else "no review"        
        no_of_reviews.append(number_of_reviews)
        
        no_of_likes = info.find('div', class_="listing-hotels-likes").text.replace(" ", "")[:-len('peoplelikethishotel')].replace(",", "")        
        likes.append(no_of_likes)
    
```

```python
hotels = pd.DataFrame({"Hotel name":name, 'Address': address, "Price (avg/night)":price_per_night, "Facilities":facilities, 
                      "Rating":rating, "Reviews":no_of_reviews, "Likes":likes })
```
